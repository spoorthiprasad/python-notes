```python
from sklearn import datasets    #import library
```

```python
iris=datasets.load_iris()    #import data set
```

```python
iris
```

```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)    # feature_names contain all length(spel and petal) and width(sepel,petal)
```

```python
iris.target_names
```

```python
group_1=df[df['target']==0]['sepal width (cm)']
group_2=df[df['target']==1]['sepal width (cm)']
group_3=df[df['target']==2]['sepal width (cm)']
```

```python
group_1.mean(),group_2.mean(),group_3.mean()
```

```python
import seaborn as sns
sns.displot(data=df,
            y="sepal width",
           
            fill=True,
            kind="kde",
            pallette=['blur','orange','green'])
            
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.scatter(range(1,51),group_1)
plt.scatter(range(1,51),group_2)
plt.scatter(range(1,51),group_3)
plt.legend(iris.target_names)
plt.show()
```

```python
from scipy import stats
stats.f_oneway(group_1,group_2,group_3)
```

```python
df1=pd.DataFrame(iris.data)
stats.f_oneway(df1.iloc[:,0],df1.iloc[:,1],df1.iloc[:,2],df1.iloc[:,3])
```

```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data=[[14,4],
      [0,10]]   # if u want to create an list inside a list then go to dataframe
data
```

```python
df=pd.DataFrame(data,
                index=['athlete','non athlete'],  #row name
                columns=['non smoker','smoker'])  #column name
```

```python
df
```

```python
df.plot(kind='bar',
        stacked=False)
```

```python
chi2,p,dof,exp=chi2_contingency(data)
```

```python
chi2_contingency(data)
```
