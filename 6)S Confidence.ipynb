```python
#find confidence interval first using normal distribution
#here stats.norm.interval() will give you directly conf.interval for 95% confidence
g_ci=stats.norm.interval(0.95,
loc=g.gain.mean(),
scale=g.gain.std())
print(g_ci)
print("mean at 95% confidence interval is ",np.round(g_ci,4))
#           #np is numpy
#for glaxo max profit or loss can be expected betwwen -2.5% to 2.6% on a paticular
```

```python
g_ci
```

```python
b_ci=stats.norm.interval(0.95,   #0.95 cofidence level
loc=b.gain.mean(),
scale=b.gain.std())
print(b_ci)
```

```python
print("mean of 95% confidence interval :",np.round(b_ci,4))   # range between -5.1 to 5.2
```

```python
#confidence interval if we donot know an sigma (t distribution)
g_ci=stats.t.interval(0.95,1737,          #1737_ degree of freedom should be given in case of t distribution
                         loc=g.gain.mean(),
                         scale=g.gain.std())
g_ci
```

```python
b_ci=stats.t.interval(0.95,1737,        # degree of freedom always equal to number of rows -1(n-1)
loc=b.gain.mean(),scale=b.gain.std())
b_ci
```
