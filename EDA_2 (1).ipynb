```python
df['Ozone'].quantile(0.25)  #lower quantile at 25%
```

```python
df['Ozone'].quantile(0.75)    #upper quantile at 75%
```

```python
#outlier detection :user defined function to calculate upper and lower extreme
def outlier_detection(data,colname):
  q1=data[colname].quantile(0.25)
  q2=data[colname].quantile(0.75)
  iqr=q2-q1
  lower_extreme = q1-(1.5*iqr)
  upper_extreme = q2+(1.5*iqr)
  return lower_extreme , upper_extreme, q2,q1
```

```python
outlier_detection(df,"Ozone")  #more than all value are considered as outliers
```

```python
outlier_detection(df,"Wind")
```
